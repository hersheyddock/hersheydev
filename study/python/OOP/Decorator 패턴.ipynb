## Decorator pattern
- ë°ì½”ë ˆì´í„° íŒ¨í„´ì˜ êµ¬ì¡° 

1) ìƒìœ„ ê³„ì¸µ : ë°ì½”ë ˆì´íŒ…  
    - í•˜ìœ„ ê³„ì¸µ 1.1) decorator contetns  
    - í•˜ìœ„ ê³„ì¸µ 1.2) functions to apply decorator (Target)

```python
#1.ìƒìœ„ í•¨ìˆ˜
def copyright(func):
    
    def new_func():
        print("@copyright") # decorator contents
        func() # ë°ì½”ë ˆì´íŒ… íŒ¨í„´ ì ìš©ì˜ Target í•¨ìˆ˜
        
    return new_func

@copyright
def smile():
    print("ğŸ¤£")

@copyright
def angry():
    print("ğŸ¤¯")

@copyright    
def love():
    print("ğŸ¥°")

@copyright    
def fire():
    print("ğŸ”¥")
    
@copyright    
def check():
    print("âœ…")

@copyright    
def star():
    print("ğŸ¤©")

@copyright    
def sleepy():
    print("ğŸ˜ª")

@copyright    
def devil():
    print("ğŸ˜ˆ")

@copyright    
def hi():
    print("ğŸ‘‹")
```
